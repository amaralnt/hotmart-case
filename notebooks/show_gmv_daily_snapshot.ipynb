```python
# Imports
import pandas as pd
import pyarrow.parquet as pq
import glob
import os
```

```python
# Reading the parquet ETL output
parquet_dir = '../data/curated/gmv_daily_snapshot/'
parquet_files = glob.glob(os.path.join(parquet_dir, '*.parquet'))

if parquet_files:
    dfs = [pq.read_table(f).to_pandas() for f in parquet_files]
    df = pd.concat(dfs, ignore_index=True)
else:
    df = pd.DataFrame()
```

```python
# Checking the data
df.sort_values(['purchase_id', 'snapshot_date'])
```
